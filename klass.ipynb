```python
n = int(input("Кількість студентів: "))
students = []

for i in range(n):
    student = {
        "author": input("ПІБ: "),
        "year": int(input("Курс: ")),
        "group": int(input("Номер групи: ")),
        "geo": int(input("Оцінка з геометрії: ")),
        "alg": int(input("Оцінка з алгебри: ")),
        "hist": int(input("Оцінка з історії: ")),
        "inf": int(input("Оцінка з інформатики: ")),
        "fil": int(input("Оцінка з філософії: "))
    }
    students.append(student)

print("\nВідмінники:")
for student in students:
    if (student["geo"] >= 10 and student["alg"] >= 10 and
        student["hist"] >= 10 and student["inf"] >= 10 and
        student["fil"] >= 10):
        print(f"{student['author']}")
```

```python
class Book:
    pass

n = int(input("Кількість студентів: "))
books = []

for i in range(n):
    b = Book()
    b.author = input("ПІБ: ")
    b.year = int(input("Курс: "))
    b.group = int(input("Номер групи: "))
    b.geo = int(input("Оцінка з геометрії: "))
    b.alg = int(input("Оцінка з алгебри: "))
    b.hist = int(input("Оцінка з історії: "))
    b.inf = int(input("Оцінка з інформатики: "))
    b.fil = int(input("Оцінка з філософії: "))

    books.append(b)

print("\nВідмінники:")
for b in books:
    if (b.geo >= 10 and b.alg >= 10 and b.hist >= 10 and b.inf >= 10 and b.fil >= 10):
        print(f"{b.author}")
```
